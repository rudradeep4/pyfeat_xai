```python
import numpy as np
import pandas as pd
import cleese_stim as cleese
from cleese_stim.engines import FaceWarp
from PIL import Image
import cv2
import mediapipe as mp
import os
from os import path
import glob
import tqdm
import pickle
from mtrf import TRF
import librosa
from scipy.signal import resample as sci_resample
from scipy.ndimage import gaussian_filter1d
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
np.seterr(divide='ignore', invalid='ignore')
```

```python
def mls_rigid_deformation(vy, vx, p, q, alpha=1.0, eps=1e-8):
    """ Rigid deformation
    
    Parameters
    ----------
    vx, vy: ndarray
        coordinate grid, generated by np.meshgrid(gridX, gridY)
    p: ndarray
        an array with size [n, 2], original control points, in (y, x) formats
    q: ndarray
        an array with size [n, 2], final control points, in (y, x) formats
    alpha: float
        parameter used by weights
    eps: float
        epsilon
    
    Return
    ------
        A deformed image.
    """
    q = np.ascontiguousarray(q.astype(np.int16))
    p = np.ascontiguousarray(p.astype(np.int16))

    # Exchange p and q and hence we transform destination pixels to the corresponding source pixels.
    p, q = q, p

    grow = vx.shape[0]  # grid rows
    gcol = vx.shape[1]  # grid cols
    ctrls = p.shape[0]  # control points

    # Compute
    reshaped_p = p.reshape(ctrls, 2, 1, 1)                                              # [ctrls, 2, 1, 1]
    reshaped_v = np.vstack((vx.reshape(1, grow, gcol), vy.reshape(1, grow, gcol)))      # [2, grow, gcol]
    
    w = 1.0 / (np.sum((reshaped_p - reshaped_v).astype(np.float32) ** 2, axis=1) + eps) ** alpha    # [ctrls, grow, gcol]
    w /= np.sum(w, axis=0, keepdims=True)                                               # [ctrls, grow, gcol]

    pstar = np.zeros((2, grow, gcol), np.float32)
    for i in range(ctrls):
        pstar += w[i] * reshaped_p[i]                                                   # [2, grow, gcol]

    vpstar = reshaped_v - pstar                                                         # [2, grow, gcol]
    reshaped_vpstar = vpstar.reshape(2, 1, grow, gcol)                                  # [2, 1, grow, gcol]
    neg_vpstar_verti = vpstar[[1, 0],...]                                               # [2, grow, gcol]
    neg_vpstar_verti[1,...] = -neg_vpstar_verti[1,...]                                  
    reshaped_neg_vpstar_verti = neg_vpstar_verti.reshape(2, 1, grow, gcol)              # [2, 1, grow, gcol]
    mul_right = np.concatenate((reshaped_vpstar, reshaped_neg_vpstar_verti), axis=1)    # [2, 2, grow, gcol]
    reshaped_mul_right = mul_right.reshape(2, 2, grow, gcol)                            # [2, 2, grow, gcol]

    # Calculate q
    reshaped_q = q.reshape((ctrls, 2, 1, 1))                                            # [ctrls, 2, 1, 1]
    qstar = np.zeros((2, grow, gcol), np.float32)
    for i in range(ctrls):
        qstar += w[i] * reshaped_q[i]                                                   # [2, grow, gcol]
    
    temp = np.zeros((grow, gcol, 2), np.float32)
    for i in range(ctrls):
        phat = reshaped_p[i] - pstar                                                    # [2, grow, gcol]
        reshaped_phat = phat.reshape(1, 2, grow, gcol)                                  # [1, 2, grow, gcol]
        reshaped_w = w[i].reshape(1, 1, grow, gcol)                                     # [1, 1, grow, gcol]
        neg_phat_verti = phat[[1, 0]]                                                   # [2, grow, gcol]
        neg_phat_verti[1] = -neg_phat_verti[1]
        reshaped_neg_phat_verti = neg_phat_verti.reshape(1, 2, grow, gcol)              # [1, 2, grow, gcol]
        mul_left = np.concatenate((reshaped_phat, reshaped_neg_phat_verti), axis=0)     # [2, 2, grow, gcol]
        
        A = np.matmul((reshaped_w * mul_left).transpose(2, 3, 0, 1), 
                        reshaped_mul_right.transpose(2, 3, 0, 1))                       # [grow, gcol, 2, 2]

        qhat = reshaped_q[i] - qstar                                                    # [2, grow, gcol]
        reshaped_qhat = qhat.reshape(1, 2, grow, gcol).transpose(2, 3, 0, 1)            # [grow, gcol, 1, 2]

        # Get final image transfomer -- 3-D array
        temp += np.matmul(reshaped_qhat, A).reshape(grow, gcol, 2)                      # [grow, gcol, 2]

    temp = temp.transpose(2, 0, 1)                                                      # [2, grow, gcol]
    normed_temp = np.linalg.norm(temp, axis=0, keepdims=True)                           # [1, grow, gcol]
    normed_vpstar = np.linalg.norm(vpstar, axis=0, keepdims=True)                       # [1, grow, gcol]
    transformers = temp / normed_temp * normed_vpstar  + qstar                          # [2, grow, gcol]
    nan_mask = normed_temp[0] == 0

    # Replace nan values by interpolated values
    nan_mask_flat = np.flatnonzero(nan_mask)
    nan_mask_anti_flat = np.flatnonzero(~nan_mask)
    transformers[0][nan_mask] = np.interp(nan_mask_flat, nan_mask_anti_flat, transformers[0][~nan_mask])
    transformers[1][nan_mask] = np.interp(nan_mask_flat, nan_mask_anti_flat, transformers[1][~nan_mask])

    # Remove the points outside the border
    transformers[transformers < 0] = 0
    transformers[0][transformers[0] > grow - 1] = 0
    transformers[1][transformers[1] > gcol - 1] = 0
    
    return transformers.astype(np.int16)
    

def resample_signal(signal, length, standardize=True):
    signal_resampled = sci_resample(signal, len(np.arange(0, length, 1/30)))
    
    signal_resampled = gaussian_filter1d(signal_resampled, sigma=2)

    if standardize:
        signal_resampled = StandardScaler().fit_transform(signal_resampled.reshape(-1, 1))

    return signal_resampled.flatten()


def get_data(data_type, file, length, au=None):
    match data_type:
        case 'rms':
            audio, audio_sr = librosa.load(file)
            rms_win = 0.5
            rms_hop = 1/30
            rms = librosa.feature.rms(y=audio, frame_length=int(30*rms_win), hop_length=int(30*rms_hop))
            data=rms[0]
        case 'auditory_nerve':
            file = file.replace(os.sep, '/')
            df = pd.read_csv(file, usecols=[1])
            data = df.to_numpy().flatten()
        case 'au':
            data = pd.read_csv(file)[au].to_numpy()
    
    data_resampled = resample_signal(data, length, standardize=True)

    return data_resampled


def extract_frames(video, write_path):
    cap = cv2.VideoCapture(video)
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count/fps
    
    idx = 0
    while(True):
        cap.set(cv2.CAP_PROP_POS_MSEC,(idx*fps))
        ret, frame = cap.read()
        if ret:
            path = f'{write_path}/{str(idx)}.png'
            frame_resized = cv2.resize(frame, None, fx=0.5, fy=0.5, interpolation= cv2.INTER_LINEAR)
            cv2.imwrite(path, frame_resized)
            # cv2.imwrite(path, frame)
            idx += 1
        else:
            break
    
    cap.release()
    cv2.destroyAllWindows()
```

```python
def transform_frame(kernel, base_frame, gain, scale, dfmxy_file):
    
    img = FaceWarp.load_file(base_frame)[0]
    
    DFMXY_EXT = '.txt' # default dfm extension is .txt
    
    MLS_ALPHA = 1.2
    COVARIANCE_MAT = [[0.0001, 0.0], [0.0, 0.0001]]
    NUM_FILES = 1
    FACE_THRESH = 0.5
    
    # Load all the different sources of landmark targets
    mls_targets = []
    mls_targets += kernel.idx.to_list()
    
    # Retrieve face landmarks from Mediapipe
    results = None
    with mp.solutions.face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1, refine_landmarks=True, min_detection_confidence=FACE_THRESH) as face_mesh:
        results = face_mesh.process(img)
    
    # Retrieve targeted landmarks for Mediapipe's results
    height, width, _ = img.shape
    np_landmarks = []
    p = []

    for face_landmarks in results.multi_face_landmarks:
        # Compute pixel size of the detected face
        x = [int(lm.x * width) for lm in face_landmarks.landmark]
        y = [int(lm.y * height) for lm in face_landmarks.landmark]
        np_landmarks = np.column_stack((x, y))
        face_min = np.amin(np_landmarks, axis=0)
        face_max = np.amax(np_landmarks, axis=0)
        face_size = face_max[1] - face_min[1]
        
        # Select landmarks to be deformed
        for i, lm_idx in enumerate(mls_targets):
            pt = np_landmarks[lm_idx]
            p.append(pt)
    
    # Setup fixed anchor point on the image's corners for the MLS
    p.append([0, 0])
    p.append([width - 1, 0])
    p.append([width - 1, height - 1])
    p.append([0, height - 1])
    p = np.array(p)
    
    gridX = np.arange(0, width, dtype=np.int16)
    gridY = np.arange(0, height, dtype=np.int16)
    vy, vx = np.meshgrid(gridX, gridY)
    
    for i in range(NUM_FILES):
        mls_offsets = kernel.iloc[:, 3:].to_numpy() * gain
        # mls_offsets *= face_size
        q = p[:-4] + mls_offsets
        q = np.concatenate((q, p[-4:]))
    
        dfmxy = np.hstack((np.array([mls_targets]).T, p[:-4], mls_offsets))
        new_kernel = pd.DataFrame({
            'index': np.int_(dfmxy[:, 0]),
            'posX': np.int_(dfmxy[:, 1]),
            'posY': np.int_(dfmxy[:, 2]),
            'defX': dfmxy[:, 3],
            'defY': dfmxy[:, 4]
        })
        k = new_kernel.copy()
        # rescale
        k.defX = scale * k.defX
        k.defY = scale * k.defY
        k.set_index('index').to_csv(dfmxy_file, index_label='index', header=None)
    
        dfmxy = FaceWarp.load_dfmxy(dfmxy_file)
        transformed = cleese.process_file(
                                            FaceWarp,
                                            base_frame,
                                            './configs/mediapipe.toml',
                                            dfmxy=dfmxy
                                        )

        return transformed
```

```python
BASE_DIR = './create_gif/35_1_3'
VIDEO_ID = '35_1_3'

cap = cv2.VideoCapture(f'{BASE_DIR}/{VIDEO_ID}.mov')
fps = cap.get(cv2.CAP_PROP_FPS)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
duration = frame_count/fps

cap.release()
```

```python
stim = get_data('auditory_nerve', f'{BASE_DIR}/{VIDEO_ID}.csv',  duration)
resp = get_data('au', f'{BASE_DIR}/{VIDEO_ID}_aus.csv', duration, au='AU12')
actual_resp = MinMaxScaler(feature_range=(0, 1)).fit_transform(np.array(resp).reshape(-1, 1)).flatten()

with open('./create_gif/trf_AU12.pkl', 'rb') as f:
    trf = pickle.load(f)
    
prediction, correlation = trf.predict(stimulus=stim, response=resp)
predicted_resp = MinMaxScaler(feature_range=(0, 1)).fit_transform(np.array(prediction).reshape(-1, 1)).flatten()
# smooth for less jittery animation
predicted_resp = gaussian_filter1d(predicted_resp, sigma=2)
```

```python
plt.plot(actual_resp, label='actual')
plt.plot(predicted_resp, label='predicted')
plt.legend()
```

```python
kernel = pd.read_csv('./kernels/avgkernel_AU12.dfmxy', header=None)
kernel.columns = ['idx', 'posX', 'posY', 'defX', 'defY']
kernel = kernel.astype({'idx': int})

monalisa = pd.read_csv('./monalisa.random.dfmxy')
kernel = kernel.loc[kernel.idx.isin(monalisa.iloc[:, 0])].reset_index(drop=True)
```

```python
extract_frames(video='./create_gif/35_1_3/35_3_M.mov', write_path='./create_gif/35_1_3/frames')
```

```python
original_frames = glob.glob('./create_gif/35_1_3/frames/*.png')
processed_frames = []

for i in tqdm.tqdm(range(len(predicted_resp))):
    transformed = transform_frame(
        kernel=kernel,
        base_frame=f'./create_gif/35_1_3/frames/{i}.png',
        gain=predicted_resp[i]-actual_resp[i],
        scale=1, 
        dfmxy_file=f'./create_gif/35_1_3/frames/{i}.dfmxy'
    )
    with Image.fromarray(transformed) as im:
        im.save(f'./create_gif/35_1_3/frames/mod_{i}.png')
    
    processed_frames.append(Image.fromarray(transformed))
```
